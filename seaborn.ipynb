HeatMap






```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/museum_visitors.csv', index_col=0, parse_dates=[0], usecols=[0, 1])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['year', 'month']).max()
df2
```

```python
df2.unstack(level=1)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(data=df2.unstack(level=1), cmap='YlGnBu', linewidths=1, square=True )
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(data=df2.unstack(level=1), cmap="RdGy_r", linewidths=1, square=True)
ls = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=ls)
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel("Minutes")
plt.show()
```

```python
data = pd.melt(df, id_vars=['Month'], value_vars=['AA', 'AS', 'DL', 'WN', 'MQ'])
data
```

ustunli grafiklar catplot()


```python
sns.catplot(data=data, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```
